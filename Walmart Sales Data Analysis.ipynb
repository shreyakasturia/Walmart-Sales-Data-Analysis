```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("/Users/shreyakasturia/Downloads/WalmartSalesData.csv")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
time_of_day = []
for t in df["Time"]:
    if t>'00:00:00' and t<'12:00:00':
        time_of_day.append("Morning")
    elif t>'12:01:00' and t<'16:00:00':
        time_of_day.append("Afternoon")
    else:
        time_of_day.append("Evening")
```

```python
df['time_of_day'] = time_of_day
```

```python
df.head()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df['day_of_week'] = df['Date'].dt.day_name()
```

```python
df.head()
```

```python
df['month_name'] = df['Date'].dt.month_name()
```

```python
df.head()
```

```python
df2 = df.groupby(['Product line'])['Product line'].count().sort_values(ascending=False)
```

```python
df2.head()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
ax = df2.plot(kind='bar', edgecolor='black', color='lightblue', alpha=0.7)

plt.title('Most Selling Product Line')
plt.xlabel('Product Line')
plt.ylabel('Count')

plt.xticks(rotation='vertical')
plt.tight_layout()

# Add value labels to each bar
for i in ax.patches:
    ax.text(i.get_x() + i.get_width() / 2, i.get_height() + 0.5, str(i.get_height()), ha='center')

plt.show()
```
